```python
# ── CELL 2: Imports & Configuration ──────────────────────────────────────
# Rule: ALL imports live here. If you need a new library later,
# add it here — never inside a later cell. This makes the notebook
# easier to debug and ensures everything loads in the correct order.

import mlflow
import pandas as pd
import os

# Pin the tracking URI explicitly so MLflow always writes to the
# same location, regardless of where Jupyter was launched from.
# Replace this path with your actual project folder path.
TRACKING_URI = "sqlite:///D:/cyberlaw_proj/mlflow.db"
mlflow.set_tracking_uri(TRACKING_URI)

# The Excel file we built — Sheet 1 is our main data
DATA_FILE = "Cyberlaw_Women_India_Verified_2024_2026.xlsx"
SHEET_NAME = "Platform_Case_Data"

# Experiment name in MLflow — all runs will appear under this
EXPERIMENT_NAME = "Cyberlaw_Case_Tracker_v3"

print(f"MLflow version  : {mlflow.__version__}")
print(f"Tracking URI    : {mlflow.get_tracking_uri()}")
print(f"Data file       : {DATA_FILE}")
print("✅ Configuration loaded successfully")
```

## Step 1: Load & Validate Data
We load the verified dataset from the Excel file and perform basic 
validation — checking row count, column names, and null values — 
before logging anything to MLflow. This prevents corrupt or 
incomplete data from entering our tracking system.

```python
# ── CELL 4: Load & Validate Data ─────────────────────────────────────────
# We read Sheet 1 of the Excel file — the main platform-level case data.
# openpyxl is required to read .xlsx files with pandas.

df = pd.read_excel(DATA_FILE, sheet_name=SHEET_NAME, engine="openpyxl")

# ── Basic Validation ──
print(f"Rows loaded      : {len(df)}")
print(f"Columns found    : {list(df.columns)}")
print(f"\nNull values per column:")
print(df.isnull().sum())

# ── Preview first 5 rows ──
df.head()



# Recalculate conviction_rate_pct directly in Python.
# We use cases_convicted / cases_reported * 100 to get a percentage.
# The round() keeps it to 2 decimal places for readability.
# We guard against division by zero with a conditional —
# if cases_reported is 0 for any row, the rate becomes 0.0 instead of crashing.
df["conviction_rate_pct"] = df.apply(
    lambda row: round((row["cases_convicted"] / row["cases_reported"]) * 100, 2)
    if row["cases_reported"] > 0 else 0.0,
    axis=1
)

# Verify it worked — you should see actual percentages, not zeros
print(df[["case_id", "platform", "cases_reported", "cases_convicted", "conviction_rate_pct"]].to_string())
```

## Step 2: MLflow Experiment Setup & Logging
Each row in our dataset becomes one MLflow "run" — think of a run 
as a case file entry in a filing cabinet. Parameters hold categorical 
data (platform, state, crime category). Metrics hold numeric data 
(cases reported, chargesheeted, convicted, conviction rate). Tags 
hold reference data (source URLs, platform type).

This separation matters because MLflow can plot and compare metrics 
across runs, but not parameters — so anything we want to visualise 
as a chart inside MLflow must be a metric.

```python
# ── CELL 6: MLflow Logging ────────────────────────────────────────────────
# IMPORTANT: We only run this cell ONCE on a fresh dataset.
# Running it multiple times creates duplicate runs in MLflow.
# If you need to re-log, go to the MLflow UI, delete the experiment,
# then re-run this cell.

# Step 1: Clear all existing runs first
client = mlflow.tracking.MlflowClient()
existing_runs = client.search_runs(experiment_ids=[experiment.experiment_id])
for run in existing_runs:
    client.delete_run(run.info.run_id)
print(f"Cleared {len(existing_runs)} old runs")


mlflow.set_experiment(EXPERIMENT_NAME)

for index, row in df.iterrows():
    # run_name = case_id (e.g. P-01, P-02) — visible in MLflow UI
    with mlflow.start_run(run_name=str(row["case_id"])):

        # ── Parameters (categorical / text fields) ──────────────────
        mlflow.log_param("year",           row["year"])
        mlflow.log_param("state",          row["state"])
        mlflow.log_param("platform",       row["platform"])
        mlflow.log_param("platform_type",  row["platform_type"])
        mlflow.log_param("crime_category", row["crime_category"])
        mlflow.log_param("laws_invoked",   row["laws_invoked"])
        mlflow.log_param("case_status",    row["case_status"])
        mlflow.log_param("convicted_yn",   row["convicted_yn"])

        # ── Metrics (numeric — these can be plotted in MLflow) ───────
        mlflow.log_metric("cases_reported",      int(row["cases_reported"]))
        mlflow.log_metric("cases_chargesheeted",  int(row["cases_chargesheeted"]))
        mlflow.log_metric("cases_convicted",      int(row["cases_convicted"]))

        # conviction_rate_pct may have NaN if formula didn't recalculate
        # so we handle that safely before logging
        conv_rate = row.get("conviction_rate_pct", 0)
        mlflow.log_metric("conviction_rate_pct",
                          float(conv_rate) if pd.notna(conv_rate) else 0.0)

        # ── Tags (reference / lookup fields) ────────────────────────
        mlflow.set_tag("source",      str(row["data_source"]))
        mlflow.set_tag("source_url",  str(row["source_url"]))

print(f"✅ Logged {len(df)} runs to experiment: {EXPERIMENT_NAME}")
```

## Step 3: Verify Logged Runs
We pull the runs back from MLflow and confirm the count matches 
our dataset. We also preview 3 runs to sanity-check that parameters, 
metrics, and tags all landed correctly.

```python
# ── CELL 8: Verification ─────────────────────────────────────────────────
client = mlflow.tracking.MlflowClient()
experiment = client.get_experiment_by_name(EXPERIMENT_NAME)
runs = client.search_runs(experiment_ids=[experiment.experiment_id])

print(f"Runs in MLflow   : {len(runs)}")
print(f"Rows in DataFrame: {len(df)}")
print(f"Match            : {'✅ Yes' if len(runs) == len(df) else '❌ Mismatch — check for duplicates'}")

# Preview 3 runs
print("\n── Sample Run Preview ──")
for run in runs[:3]:
    print(f"\nRun : {run.info.run_name}")
    print(f"  Params  : {run.data.params}")
    print(f"  Metrics : {run.data.metrics}")
```

```python
# ── Verification: Count & Conviction Rates ───────────────────────────
client = mlflow.tracking.MlflowClient()
experiment = client.get_experiment_by_name("Cyberlaw_Case_Tracker_v3")
runs = client.search_runs(experiment_ids=[experiment.experiment_id])

print(f"Total runs logged : {len(runs)}")
print(f"Expected          : 32")
print(f"Status            : {'✅ Clean' if len(runs) == 32 else '❌ Mismatch'}")

# Check conviction rates are no longer zero
zero_rates = [r for r in runs if r.data.metrics.get("conviction_rate_pct", 0) == 0]
print(f"\nRuns with zero conviction rate : {len(zero_rates)}")
print(f"Status : {'✅ All good' if len(zero_rates) == 0 else '⚠️ Some zeros remain — check those rows'}")
```

```python
# ── Summary: Platform vs Conviction Rate ─────────────────────────────
summary = []
for run in runs:
    summary.append({
        "run"              : run.info.run_name,
        "platform"         : run.data.params.get("platform"),
        "cases_reported"   : run.data.metrics.get("cases_reported"),
        "cases_convicted"  : run.data.metrics.get("cases_convicted"),
        "conviction_rate"  : run.data.metrics.get("conviction_rate_pct")
    })

summary_df = pd.DataFrame(summary).sort_values("conviction_rate", ascending=False)
print(summary_df.to_string(index=False))
```
